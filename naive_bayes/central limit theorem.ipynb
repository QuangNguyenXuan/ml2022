```python
import seaborn as sns
import random
sns.set_theme()

# load image
img = cv2.imread('Sydney-Opera-House.jpg', 0)
height, width = img.shape

# ...
num_of_samples = 10000
sample_size = 50

means = []
for _ in range(num_of_samples):    
    mean = 0
    for _ in range(sample_size):
        # generate a random point
        x = random.randint(0, height-1) # remove
        y = random.randint(0, width-1)  # remove
        
        # get pixel value and add to mean
        mean = mean + img[x][y]         # remove
    # normalize to get mean
    mean = mean/sample_size
    
    # add to list
    means.append(mean)
```

```python
plot = sns.distplot(means)
fig = plot.get_figure()
fig.savefig("out.png")
```

```python
import random
import math

num_of_samples = 10000
sample_size = 30
side = 500

means = []
for _ in range(num_of_samples):    
    mean = 0
    for _ in range(sample_size):
        # generate a random point 
        x = random.randint(0, side-1)  # remove
        y = random.randint(0, side-1)  # remove
        
        # distance to (0,0)
        distance = math.sqrt(x*x + y*y)
        
        # get pixel value and add to mean
        mean = mean + distance         # remove
    # normalize to get mean
    mean = mean/sample_size
    
    # add to list
    means.append(mean)
```

```python
import seaborn as sns
sns.set_theme()

sns.distplot(means)
```
