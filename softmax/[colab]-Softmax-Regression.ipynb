#Softmax Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Sample 1**

```python
x = np.array([[1.0, 0.5]])
Y = np.array([0])
```

```python
def convert_one_hot(y, k):
    one_hot = np.zeros((len(y), k))
    one_hot[np.arange(len(y)), y] = 1
    return one_hot

n_classes = 2
Y_one_hot = convert_one_hot(Y, n_classes)
Y_one_hot
```

```python
y = Y_one_hot[0]
```

```python
x, y
```

```python
theta = np.array([[0.1, 0.2], [0.3, 0.4]])
theta
```

```python
# define softmax function
def softmax_function(z):
    return np.exp(z) / np.sum(np.exp(z))
```

```python
# compute y_hat
def predict(x, theta):
    z = np.dot(x , theta)
    y_hat = np.exp(z) / np.sum(np.exp(z))
    return z, y_hat

z, y_hat = predict(x, theta)
z, y_hat
```

```python
# compute loss
def compute_loss(y_hat, y):
    loss = -np.log(np.sum(y_hat*y, axis=1))
    return loss

loss = compute_loss(y_hat, y)
loss
```

```python
# compute gradient
def compute_gradient(y_hat, y, x):
    gradient = np.dot(x.transpose(), (y_hat - y))
    return gradient

gradient = compute_gradient(y_hat, y, x)
gradient
```

```python
# update weights
learning_rate = 0.1
theta -= (learning_rate * gradient)
theta
```

```python
# compute y_hat
def predict(x, theta):
    z = np.dot(x , theta)
    y_hat = np.exp(z) / np.sum(np.exp(z))
    return z, y_hat

z, y_hat = predict(x, theta)
z, y_hat
```

```python
# compute loss
def compute_loss(y_hat, y):
    loss = -np.log(np.sum(y_hat*y, axis=1))
    return loss

loss = compute_loss(y_hat, y)
loss
```
